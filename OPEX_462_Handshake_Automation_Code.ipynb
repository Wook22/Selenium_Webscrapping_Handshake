```python
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver

from bs4 import BeautifulSoup

import pandas as pd
import time
import logging
```

```python

def login(url, email, password):
    # Configure the WebDriver options
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')  # Optional: Run Chrome in headless mode (no GUI)
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')

    # Set up the WebDriver using the options
    driver = webdriver.Chrome(options=chrome_options)

    try:
        # Navigate to the login page
        driver.get(url)

        # Find the email input element by its ID
        email_input = driver.find_element(By.ID, 'email-address-identifier')

        # Enter the email
        email_input.send_keys(email)

        # Find the 'Next' button and click it
        next_button = driver.find_element(By.XPATH, "//button[@type='submit']")
        next_button.click()

        # Wait for the password input field to be visible
        password_input = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.ID, 'password'))
        )

        # Enter the password
        password_input.send_keys(password)

        # Find the 'Login' or 'Submit' button and click it
        sign_in_button = driver.find_element(By.XPATH, "//button[contains(text(), 'Sign In')]")
        sign_in_button.click()

        # Check if the close button for the popup is present
        try:
            # Wait for the close button to be present, if not raise TimeoutException
            close_button = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.XPATH, "//button[@class='close']"))
            )

            # Click the close button
            close_button.click()

            # Wait for the close button to be present again
            close_button = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.XPATH, "//button[@class='close']"))
            )

            # Click the close button again
            close_button.click()

        except TimeoutException:
            # The close button is not present, or the popup did not appear, continue with the script
            pass

        # Now, proceed to click on the "Talent" element
        talent_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'a[data-hook="nav-item-talent"] span.sc-erijoS.eufRUy'))
        )

        talent_element.click()

        # Perform additional actions or scraping as needed

        # Navigate to the Talent page by directly modifying the URL
        # "Just for now since going to talent page is not working"
        talent_page_url = "https://app.joinhandshake.com/emp/talent"
        driver.get(talent_page_url)

        return driver

    except Exception as e:
        logging.error(f"Error during login: {e}")
        return None  # Return None to indicate failure

```

```python
# Filter the list

def filter_school(driver, school_names):
    try:
        # Click on the 'School' button
        school_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//button[contains(text(), 'School')]"))
        )
        school_button.click()

        # Split the school_names string into a list
        school_names_list = [name.strip() for name in school_names.split(',')]

        for school_name in school_names_list:
            # Select the specified school from the list
            school_option = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, f"//li[contains(text(), '{school_name}')]"))
            )
            school_option.click()

        # Click on the 'Apply filters' button
        apply_filters_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//button[contains(@class, 'sc-eGugkK fmxlbE')]"))
        )
        apply_filters_button.click()

        # Now, the filters are applied, and you can proceed with further actions or scraping

    except Exception as e:
        print(f"Error during school filtering: {e}")

    finally:
        return driver
```

```python
def extract_data(driver):
    # Check if the close button for the popup is present
    try:
        # Wait for the close button to be present, if not raise TimeoutException
        close_button = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.XPATH, "//button[@class='close']"))
        )

        # Click the close button
        close_button.click()

        # Wait for the close button to be present again
        close_button = WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.XPATH, "//button[@class='close']"))
        )

        # Click the close button again
        close_button.click()

    except TimeoutException:
        # The close button is not present, or the popup did not appear, continue with the script
        pass

    # Initialize the DataFrame
    final_df = pd.DataFrame()
    page_number = 1  # Starting page number
    next_page_xpath = "//button[@aria-label='next page' and @aria-disabled='false']"

    while True:
        print(f"Extracting data from page {page_number}...")

        # Assuming 'html_content' is the HTML content of the page
        html_content = driver.page_source

        # Create a BeautifulSoup object
        soup = BeautifulSoup(html_content, "html.parser")

        # Check if the 'tbody' element is present
        tbody_element = soup.find("tbody")

        # Check if tbody is found
        if tbody_element:
            # Find all rows within the tbody
            rows = tbody_element.find_all("tr")

            # Initialize lists to store data
            names = []
            active_statuses = []
            degrees = []
            educations = []
            skills = []

            # Loop through each row to extract data
            for row in rows:
                # Extract information from each row
                name = row.find("button", class_="sc-iMGJAB kNiNOz")
                active_status = row.find("td", class_="sc-iKuAEz kUCFZv")
                degree = row.find("td", class_="sc-eTZCzI kDHbZN")
                education_skills = row.find("td", class_="sc-bbSSRB gsWocU")

                # Initialize here
                education = ""
                skills_text = ""
                skills_list = ""

                # Extracting Education and Skills
                if education_skills:
                    education_tag = education_skills.find("dt", string="Education:")
                    education = (
                        education_tag.find_next("dd").text.strip()
                        if education_tag
                        else ""
                    )

                    skills_tag = education_skills.find("dt", string="Skills:")
                    skills_text = (
                        skills_tag.find_next("dd").text.strip() if skills_tag else ""
                    )

                    # Remove the "+10 more" part
                    skills_text = skills_text.split("+")[0].strip()

                    # If there are links, extract them
                    skills_list = (
                        [
                            skill.text.strip()
                            for skill in education_skills.find("dt", string="Skills:")
                            .find_next("dd")
                            .find_all("a")
                        ]
                        if skills_tag
                        else []
                    )

                # Append data to lists
                names.append(name.text.strip() if name else "")
                active_statuses.append(
                    active_status.text.strip() if active_status else ""
                )
                degrees.append(degree.text.strip() if degree else "")
                educations.append(education)
                skills.append(
                    ", ".join([skills_text] + skills_list)
                    if skills_text or skills_list
                    else ""
                )

            # Create a DataFrame
            data = {
                "Name": names,
                "Active Status": active_statuses,
                "Degree": degrees,
                "Education": educations,
                "Skills": skills,
            }

            current_page_df = pd.DataFrame(data)

            # Append data to the final DataFrame
            final_df = pd.concat([final_df, current_page_df], ignore_index=True)

            print("Data extracted successfully.")
        else:
            print(f"No rows found on page {page_number}.")

        # Check if there is a next page button
        try:
            next_page = WebDriverWait(driver, 20).until(
                EC.element_to_be_clickable((By.XPATH, next_page_xpath))
            )
            # Click the next button
            print(f"Currently have {len(final_df)} rows...")
            next_page.click()
            WebDriverWait(driver, 20).until(EC.staleness_of(next_page))
            time.sleep(5)
            page_number += 1  # Increment the page number

        except TimeoutException:
            print("No more pages. Exiting the loop.")
            break

    return final_df
```

```python
# Login and get the driver
driver = login(search_url, account_email, account_password)

if driver:
    # Continue with additional actions or scraping using the 'driver'
    pass
else:
    print("Login failed.")

# Call the filter_school function only if the user wants to apply filters
if apply_filters == 'yes':
  # Prompt the user to enter multiple school names separated by a comma
  school_names_input = input("Enter the School Names for specific filtering (separated by commas): ")

  filter_school(driver, school_names_input)

# Call the extract_data function
final_df = extract_data(driver)
```
